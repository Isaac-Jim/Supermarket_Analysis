```python


# Dictionary for day mapping
day_mapping = {
    'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5,
    'Saturday': 6, 'Sunday': 7
}

# Create a new column 'day_num' by mapping the 'Days' column
df['day_num'] = df['Days'].map(day_mapping)

# Sort the DataFrame based on 'day_num'
df_sorted = df.sort_values('day_num')

# Create a line plot with proper order
plt.figure(figsize=(10,6))
sns.lineplot(x='Days', y='Total', data=df_sorted, sort=False,hue='City',errorbar=('ci', False))
plt.title('trend of total amount per day')
# Show the plot
plt.show()
```
